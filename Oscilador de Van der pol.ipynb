## Oscilador de Van der pol

```python
def VanderPol(x,t,mu):
    return [x[1],mu*(1-x[0]**2)*x[1]-x[0]]
```

```python
t = linspace (0,1000,100001)
```

```python
xout = odeint(VanderPol, [1.0, 0.0],t ,args=(0.1,))
```

```python
def fft_w(x,ffx):
    n=len(ffx)
    delta_x = x[1] - x[0]
    omega= linspace(0,1.0/delta_x, n/2 + 1)
    omega= omega*pi
    T=np.fft.rfft(ffx)/n
    espectro= abs (T)
    return omega, espectro
```

```python
omega, espectro = fft_w(t,xout[:,0])
```
