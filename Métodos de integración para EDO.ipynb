## Método de Euler 

Si tenemos un problema del tipo $y'(x)=f(x,y(x))$ con $y(x_0) = y_0$. Entonces una forma de encontrar la solución es:
$$y_{i+1} = hf(y_i , x_i) + y_i$$
para valores de $i = 1,2,3,...$ y $h=x_1 - x_0$ es el paso de integración.

````python
from pylab import *
%matplotlib inline
````

````python
def mi_euler(val_iniciales,tiempo,funcion):
    h = tiempo[1] - tiempo[0]
    solucion = zeros(len(tiempo))
    solucion[0] = val_iniciales
    for i in range(len(tiempo)-1):
        solucion[i+1] = h*funcion(solucion[i],tiempo[i]) + solucion[i]
    return array(solucion)
````

La forma de poner $y'=y$ la ecuación diferencial esta representada por es una función $f(y,x)=y$

````python
def f1(y,x):
    return y
````

````python
t = linspace(0,5,500)
y_de_x = mi_euler(1.,t,f1)
plot(t,y_de_x,color="red")
plot(t,exp(t),color="green")
show()
````

````python
def f2(y,x):
    return y*(1-y)
````

````python
t = linspace(0,10,100)
muchas_condiciones = linspace(0.1,3,30)
for cond in muchas_condiciones:
    Pob = mi_euler(cond,t,f2)
    plot(t,Pob)
show()
````

## Método de Runge-Kutta

El método de Euler no es muy eficiente y pierde precisión rápidamente ya que es de primer orden. Los matemáticos C. Runge y M. W. Kutta desarrollaron un método mucho mejor para integrar ecuaciones diferenciales.

Uno de los métodos numéricos más utilizado y desarrollado por los autores anteriores es el **Método de Runge-Kutta de cuarto orden**. Éste resuelve el problema de una ecuación diferencial:
$$
\frac{d \vec x}{dt} = f(t,\vec x)
$$
con valor inicial $\vec x(0)=\vec x_0$. 

El método de Runge-Kutta consiste en encontrar el punto $x_{n+1}$ a partir del punto inicial en pasos de $h$. Para encontrar dicho punto es necesario hacer interpolaciones de la siguiente manera:
* $k_1 = f(t_n,x_n)$
* $k_2 = f(t_n + \tfrac{h}{2} , x_n + \tfrac{h}{2}k_1)$
* $k_3 = f(t_n + \tfrac{h}{2} , x_n + \tfrac{h}{2}k_2)$
* $k_4 = f(t_n + h,x_n+hk_3)$

Note que estas interpolaciones son iteradas, es decir, $k_2$ depende de $k_1$, $k_3$ depende de $k_2$ y $k_4$ depende de $k_3$. Una vez que se tienen estas interpolaciones, el punto $x_{n+1}$ queda de la siguiente manera:
$$
x_{n+1} = x_n + \frac{h}{6}(k_1 + 2k_2 + 2 k_3 + k_4)
$$

El error que se comete con este algoritmo es de orden $O(h^5)$ y de ahí el nombre de 4º orden.

### Método de Runge Kutta de 4° orden

la función que programaremos la llamaremos ```RK``` y sus argumentos de entrada son:

* ```x_ini``` es la condición inicial y debe ser una lista o array.
* ```int_t``` es el intervalo de tiempo sobre el que se hace la integración. Debe tener dos elementos $t_{inicial} \to$ ```int_t[0]``` y $t_{final} \to$ ```int_t[1]```.
* ```funcion``` es la función que representa la ecuación diferencial. Note que ésta debe tener al menos dos parámetros: ```t``` y ```x```.
* ```h=0.1``` es el paso de integración y por defecto es $0.1$.
* ```args=[]``` son los argumentos opcionales en los que se pueden pasar parámetros a la ecuación diferencial en ```funcion```.

Las interpolaciones quedan determinadas por el código:
```python
k1 = funcion( ti , x_i ) # interpolacion k1
k2 = funcion( ti + 0.5*h , x_i + 0.5*h*k1 ) # interpolacion k2
k3 = funcion( ti + 0.5*h , x_i + 0.5*h*k2 ) # interpolacion k3
k4 = funcion( ti + h , x_i + h*k3 ) # interpolacion k4
```

Note que en la función de abajo se revisa si hay argumentos para pasar a la función que representa la ecuación diferencial. Finalmente el punto $x_{n+1}$ se calcula con la siguiente línea:
```python
x_i = x_i + (h/6.)*( k1 + 2*(k2+k3) + k4 ) # calculo de x_n+1
```

Al igual como se hizo con la función del método de Euler se tiene un pequeño código que acomoda los datos de salida para que sean más manejables. De esta manera la función ```RK``` queda de la siguiente manera:

````python
def RK(x_ini, int_t, funcion, h=0.1 , args=[]):
    # La funcion a integrar debe regresar un array
    t = arange(int_t[0],int_t[1]+h,h) # variable del intervalo de integración
    x_i = array(x_ini)
    solucion = [x_i] # variable para guardar la solucion
    for ti in t:
        if len(args) == 0:
            k1 = funcion( ti , x_i ) # interpolacion k1
            k2 = funcion( ti + 0.5*h , x_i + 0.5*h*k1 ) # interpolacion k2
            k3 = funcion( ti + 0.5*h , x_i + 0.5*h*k2 ) # interpolacion k3
            k4 = funcion( ti + h , x_i + h*k3 ) # interpolacion k4
        else:
            k1 = funcion( ti , x_i , args )
            k2 = funcion( ti + 0.5*h , x_i + 0.5*h*k1 , args )
            k3 = funcion( ti + 0.5*h , x_i + 0.5*h*k2 , args )
            k4 = funcion( ti + h , x_i + h*k3 , args )
        x_i = x_i + (h/6.)*( k1 + 2*(k2+k3) + k4 ) # calculo de x_n+1
        solucion.append(x_i)
    salida = [t]
    # Acomodando los resultados
    for i in range(len(x_ini)):
        x_i_sol = []
        for j in range(len(t)):
            x_i_sol.append(solucion[j][i])
        salida.append(x_i_sol)
    return salida
````

### El oscilador armónico como un sistema de ecuaciones de primer orden

Recordemos que la ecuación para el oscilador armónico en una dimensión está dado por:

$$ \frac{d^{2}x}{dt^2} + \omega^2 x = 0 ,$$

podemos escribir esta ecuación como un sistema de ecuaciones de primer orden:

$$ \frac{dx}{dt} = y,  \;\;\;\;\;\; \frac{dy}{dt} = -\omega ^2 x .$$

donde se ha hecho el cambio de variable $y = \frac{dx}{dt}$. Al igual que en el *notebook* del Método de Euler definimos la ecuación diferencial del oscilador armónico de la siguiente manera:

````python
def osc_armonico(t,x,omega_2=[1] ):
    #x[0] representa y, x[1] representa dy o u  
    return array( [x[1] , -omega_2[0]*x[0]] )
````

Con las siguientes líneas encontramos la solución numérica utilizando el método de Runge-Kutta y la comparamos con la solución analítica es $\sin (x)$.

````python
sol = RK( [0,1] , [0,3*pi] , osc_armonico )
plot(sol[0],sol[1],'+',label='$y(t)$ numerica')
plot(sol[0],sin(sol[0]) , label='$y(t)$ numerica')
legend(loc=4)
````

Como se puede observar en este caso basta con tener el paso $h=0.1$ y la solución numérica es prácticamente a la analítica. Esta ventaja hace que se puedan hacer integraciones mucho mejore con menos pasos, lo que hace que se ahorre mucho tiempo de cómputo. Invitamos al lector a revisar el *notebook* sobre el Método de Euler y verificar que para dicho método se necesita un paso de integración de $h=0.001$ para obtener el mismo resultado.

A continuación se muestra el espacio fase con las mismas condiciones a como se hizo con el Método de Euler para efectos comparativos.

````python
condiciones_iniciales=[[0,0.5],[0,1],[0,1.5],[0,2]]
figure(figsize=(7,7), dpi=100)
for ci in condiciones_iniciales:
    sol = RK( ci , [0,4*pi] , osc_armonico , args=[1.5] )
    plot(sol[1],sol[2] ,label=str(ci) )
title("Espacio fase"),xlabel("$y(t)$"),ylabel("$y'(t)$")
xlim([-2.5,2.5]),ylim([-2.5,2.5])
legend(loc=4),grid(True),axes().set_aspect('equal')
show()
````
