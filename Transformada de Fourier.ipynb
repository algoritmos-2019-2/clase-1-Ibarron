Se grafica con la frecuencia, si te tomas el inverso del tiempo te tomas una base para tu frecuencia partida por todo lso pedazos

```python
def fft_w(x,ffx):
    n=len(ffx)
    delta_x = x[1] - x[0]
    omega= linspace(0,1.0/delta_x, n/2 + 1)
    omega= omega*pi
    T=np.fft.rfft(ffx)/n
    espectro= abs (T)
    return omega, espectro
    
```

```python
omega, I= fft_w(t,f1(t))
```

```python
plot (omega,I)
xlim (0,6)
```
